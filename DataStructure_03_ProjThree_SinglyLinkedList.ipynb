```python
class Node:

    def __init__(self, data):
        self.data = data
        self.ref = None

class SinglyLinkedList:

    def __init__(self):
        self.head = None

    def print_linked_list(self):
        n = self.head
        if n is None:
            print("The linked list is empty!")
            return None
        else:
            while n is not None:
                print(n.data, "->", end = " ") if n.ref != None else print(n.data)
                n = n.ref
            return None

    def get_size(self):
        n = self.head
        count = 0
        while n:
            count += 1
            n = n.ref
        return count

    def max_value(self):
        n = self.head
        if not n:
            print("The linked list is empty!")
            return None
        if not n.ref:
            return n.data
        else:
            max_value = n.data
            while n:
                max_value = max(n.data, max_value)
                n = n.ref
        return max_value

    def min_value(self):
        n = self.head
        if not n:
            print("The linked list is empty!")
            return None
        if not n.ref:
            return n.data
        else:
            min_value = n.data
            while n:
                min_value = min(n.data, min_value)
                n = n.ref
        return min_value

    def get_index(self, x):
        n = self.head
        if n is None:
            print("Linked list is empty!")
        index = 0
        while n:
            if n.data==x:
                return index
            index += 1
            n = n.ref
        print("{} does not exist in the linked list!".format(x))
        return None

    def insert_empty(self, data):
        new_node = Node(data)
        n = self.head
        if n:
            print("The linked list is not empty!")
            return None
        else:
            self.head = new_node
            return None

    def add_begin(self, data):
        new_node = Node(data)
        n = self.head
        if n is None:
            return self.insert_empty(data)
        else:
            new_node.ref = self.head
            self.head = new_node
            return None

    def add_end(self, data):
        new_node = Node(data)
        n = self.head
        if n is None:
            return self.insert_empty(data)
        else:
            while n.ref:
                n = n.ref
            n.ref = new_node

    def add_after(self, data, x):
        new_node = Node(data)
        n = self.head
        if n is None:
            print("The linked list is empty!")
            return None
        else:
            while n.ref and n.data != x:
                n = n.ref
            if n.ref and n.data == x:
                new_node.ref = n.ref
                n.ref = new_node
            elif n.ref is None and n.data == x:
                return self.add_end(data)
            else:
                print("{} does not exist in the linked list!".format(x))
                return None

    def add_before(self, data, x):
        new_node = Node(data)
        n = self.head
        if n is None:
            print("The linked list is empty!")
            return None
        elif n.data == x:
            return self.add_begin(data)
        else:
            while n.ref and n.ref.data != x:
                n = n.ref
            if n.ref.data == x:
                return self.add_after(data, n.data)
            else:
                print("{} does not exist in the linked list!".format(x))
                return None

    def add_at_index(self, data, index):
        n = self.head
        length = self.get_size()
        count = 0
        if index > length:
            print("The index is out of the range!")
            return None
        if index == 0:
            return self.add_begin(data)
        elif index == length:
            return self.add_end(data)
        else:
            while count != index:
                n = n.ref
                count += 1
            return self.add_before(data, n.data)

    def delete_begin(self):
        n = self.head
        if n is None:
            print("The linked list is already empty!")
            return None
        self.head = n.ref
        return None

    def delete_end(self):
        n = self.head
        if n is None:
            print("The linked list is already empty!")
            return None
        if n.ref is None:
            return delete_begin()
        while n.ref.ref:
                n = n.ref
        n.ref = None
        return None

    def delete_by_value(self, x):
        n = self.head
        if n is None:
            print("The linked list is already empty!")
            return None
        if n.data == x:
            return self.delete_begin()
        while n.ref.ref and n.ref.data != x:
                n = n.ref
        if n.ref.ref and n.ref.data == x:
            n.ref = n.ref.ref
        elif not n.ref.ref and n.ref.data == x:
            return self.delete_end()
        else:
            print("{} does not exist in the linked list!".format(x))
            return None

    def delete_by_index(self, index):
        n = self.head
        length = self.get_size()
        count = 0
        if index >= length:
            print("The index is out of the range!")
            return None
        while count != index:
            n = n.ref
            count += 1
        return self.delete_by_value(n.data)

    def reverse_linked_list(self):
        prev_node = None
        curr_node = self.head
        while curr_node:
            next_node = curr_node.ref
            curr_node.ref = prev_node
            prev_node = curr_node
            curr_node = next_node
        self.head = prev_node

    def clear(self):
        self.head = None
        return None

#     def sort_linked_list(self):
#         n = self.head
#         if n is None:
#             print("The linked list is already empty!")
#             return None
#         if n.ref is None:
#             return n.data
#         else:
#             min_val = self.min_value()
#             self.delete_by_value(min_val)
#             return self.add_begin(min_val)+self.sort_linked_list()
#     def add_digits


LL = SinglyLinkedList()
LL.add_begin(-5)
LL.add_begin(-3)
LL.insert_empty(10)
LL.add_begin(5)
LL.add_begin(0)
LL.add_end(20)
LL.add_end(-20)
LL.add_end(30)
LL.add_after(25,20)
LL.add_before(23,25)
LL.delete_begin()
LL.delete_end()
LL.delete_by_value(23)
LL.delete_by_index(1)
LL.add_at_index(33, 3)
LL.get_index(36)
LL.print_linked_list()
LL.reverse_linked_list()
LL.print_linked_list()
print("Length of the linked list is {}".format(LL.get_size()))
print("Maximum value in the linked list is {}".format(LL.max_value()))
print("Minimum value in the linked list is {}".format(LL.min_value()))
```
