```python
def count(node):

    if node is None:
        return 0
    return 1 + count(node.lchild) + count(node.rchild)

class BinarySearchTree:

    def __init__(self, key):

        self.key = key
        self.lchild = None
        self.rchild = None

    def insert_node(self, data):

        if self.key is None:
            self.key = data
            return None
        if data == self.key:
            return None
        if data < self.key:
            if self.lchild:
                self.lchild.insert_node(data)
            else:
                self.lchild = BinarySearchTree(data)
        else:
            if self.rchild:
                self.rchild.insert_node(data)
            else:
                self.rchild = BinarySearchTree(data)
        return None

    def search_node(self, data):

        if self.key is None:
            print("Empty tree!")
            return None
        if self.key == data:
            print("Node exists!")
            return None
        if data < self.key:
            if self.lchild:
                self.lchild.search_node(data)
            else:
                print("Node does not exist!")
        else:
            if self.rchild:
                self.rchild.search_node(data)
            else:
                print("Node does not exist!")
        return None

    def preorder(self):

        print(self.key, end=" ")
        if self.lchild:
            self.lchild.preorder()
        if self.rchild:
            self.rchild.preorder()
        return None

    def inorder(self):

        if self.lchild:
            self.lchild.inorder()
        print(self.key, end=" ")
        if self.rchild:
            self.rchild.inorder()
        return None

    def postorder(self):

        if self.lchild:
            self.lchild.postorder()
        if self.rchild:
            self.rchild.postorder()
        print(self.key, end=" ")
        return None

    def min_node(self):

        curr_obj = self
        while curr_obj.lchild:
            curr_obj = curr_obj.lchild
        return curr_obj.key

    def max_node(self):

        curr_obj = self
        while curr_obj.rchild:
            curr_obj = curr_obj.rchild
        return curr_obj.key

    def delete_node(self, data, curr):

        if self.key is None:
            print("Tree is empty!")
            return None
        if data < self.key:
            if self.lchild:
                self.lchild = self.lchild.delete_node(data, curr)
            else:
                print("Node does not exist!")
        elif data > self.key:
            if self.rchild:
                self.rchild = self.rchild.delete_node(data, curr)
            else:
                print("Node does not exist!")
        else:
            if self.lchild is None:
                temp = self.rchild
                if data == curr:
                    self.key = temp.key
                    self.lchild = temp.lchild
                    self.rchild = temp.rchild
                    temp = None
                    return None
                self = None
                return temp
            if self.rchild is None:
                temp = self.lchild
                if data == curr:
                    self.key = temp.key
                    self.lchild = temp.lchild
                    self.rchild = temp.rchild
                    temp = None
                    return None
                self = None
                return temp
            node = self.rchild
            while node.lchild:
                node = node.lchild
            self.key = node.key
            self.rchild = self.rchild.delete_node(node.key, curr)
        return self

BST = BinarySearchTree(None)
BST = BinarySearchTree(10)
lst = [_ for _ in range(10)]
for elm in lst:
    BST.insert_node(elm)
BST.preorder()
if count(BST) > 1:
    BST.delete_node(10, BST.key)
else:
    print("Can't perform deletion!")
print("\n")
BST.preorder()
```

```python
# 1
# Balanced Binary Tree - Leetcode 110 - Python
class TreeNode:

    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def is_balanced(self, root: Optional[TreeNode]) -> bool:

        def dfs(node):
            if not node: return [True, 0]
            left, right = dfs(node.left), dfs(node.right)
            balanced = (left[0] and right[0] and abs(left[1] - right[1] <= 1))
            return [balanced, 1 + max(left[1], right[1])]

        return dfs(root)[0]

def test_is_balanced():
    solution = Solution()

    # Test Case 1: A balanced tree
    root1 = TreeNode(1)
    root1.left = TreeNode(2)
    root1.right = TreeNode(3)
    root1.left.left = TreeNode(4)
    root1.left.right = TreeNode(5)
    root1.right.left = TreeNode(6)
    root1.right.right = TreeNode(7)
    assert solution.is_balanced(root1) == True

    # Test Case 2: An unbalanced tree
    root2 = TreeNode(1)
    root2.left = TreeNode(2)
    root2.right = TreeNode(3)
    root2.left.left = TreeNode(4)
    root2.left.right = TreeNode(5)
    root2.left.left.left = TreeNode(6)
    assert solution.is_balanced(root2) == False

    # Test Case 3: An empty tree
    assert solution.is_balanced(None) == True

    # Test Case 4: A tree with a single node
    root3 = TreeNode(1)
    assert solution.is_balanced(root3) == True

    # Test Case 5: A tree with two levels
    root4 = TreeNode(1)
    root4.left = TreeNode(2)
    assert solution.is_balanced(root4) == True

    # Test Case 6: A tree with three levels
    root5 = TreeNode(1)
    root5.left = TreeNode(2)
    root5.right = TreeNode(3)
    root5.left.left = TreeNode(4)
    root5.left.right = TreeNode(5)
    assert solution.is_balanced(root5) == True

    # Test Case 7: A tree with varying heights
    root6 = TreeNode(1)
    root6.left = TreeNode(2)
    root6.right = TreeNode(3)
    root6.left.left = TreeNode(4)
    root6.left.right = TreeNode(5)
    root6.left.left.left = TreeNode(6)
    root6.left.left.left.left = TreeNode(7)
    assert solution.is_balanced(root6) == False

    print("All test cases passed!")

test_is_balanced()
```

```python
# 2
# Microsoft's Most Asked Question 2021 - Count Good Nodes in a Binary Tree - Leetcode 1448 - Python
class NodeTree:

    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def count_good_nodes(self, root: Optional[NodeTree]) -> int:

        def dfs(node: Optional[NodeTree], max_val: int):

            if not node: return 0

            res = 1 if node.val >= max_val else 0
            max_val = max(max_val, node.val)
            res += dfs(node.left, max_val)
            res += dfs(node.right, max_val)
            return res

        return dfs(root, root.val)

def test_count_good_nodes():
    solution = Solution()

    # Test Case 1: A tree with all nodes being good
    root1 = NodeTree(3)
    root1.left = NodeTree(1)
    root1.right = NodeTree(4)
    root1.left.left = NodeTree(3)
    root1.right.left = NodeTree(1)
    root1.right.right = NodeTree(5)
    assert solution.count_good_nodes(root1) == 4

    # Test Case 2: A tree with only the root being good
    root2 = NodeTree(5)
    root2.left = NodeTree(4)
    root2.right = NodeTree(3)
    root2.left.left = NodeTree(2)
    root2.left.right = NodeTree(1)
    root2.right.left = NodeTree(0)
    assert solution.count_good_nodes(root2) == 1

    # Test Case 3: A tree with a single node
    root3 = NodeTree(1)
    assert solution.count_good_nodes(root3) == 1

    # Test Case 4: A tree with two levels
    root4 = NodeTree(3)
    root4.left = NodeTree(1)
    root4.right = NodeTree(5)
    assert solution.count_good_nodes(root4) == 2

    # Test Case 5: A tree with three levels
    root5 = NodeTree(10)
    root5.left = NodeTree(5)
    root5.right = NodeTree(15)
    root5.left.left = NodeTree(3)
    root5.left.right = NodeTree(8)
    root5.right.left = NodeTree(12)
    root5.right.right = NodeTree(20)
    assert solution.count_good_nodes(root5) == 3

    print("All test cases passed!")

test_count_good_nodes()
```

```python
# 3
# Invert Binary Tree - Depth First Search - Leetcode 226
class NodeTree:

    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def invert_tree(self, root: NodeTree) -> NodeTree:

        if not root: return None
        tmp = root.left
        root.left = root.right
        root.right = tmp
        self.invert_tree(root.left)
        self.invert_tree(root.right)
        return root
def test_invert_tree():
    solution = Solution()

    # Test Case 1: A balanced tree
    root1 = NodeTree(1)
    root1.left = NodeTree(2)
    root1.right = NodeTree(3)
    root1.left.left = NodeTree(4)
    root1.left.right = NodeTree(5)
    root1.right.left = NodeTree(6)
    root1.right.right = NodeTree(7)

    inverted_root1 = solution.invert_tree(root1)
    assert inverted_root1.val == 1
    assert inverted_root1.left.val == 3
    assert inverted_root1.right.val == 2
    assert inverted_root1.left.left.val == 7
    assert inverted_root1.left.right.val == 6
    assert inverted_root1.right.left.val == 5
    assert inverted_root1.right.right.val == 4

    # Test Case 2: A tree with only the root
    root2 = NodeTree(5)

    inverted_root2 = solution.invert_tree(root2)
    assert inverted_root2.val == 5
    assert inverted_root2.left is None
    assert inverted_root2.right is None

    # Test Case 3: An empty tree
    assert solution.invert_tree(None) is None

    # Test Case 4: A tree with a single node
    root3 = NodeTree(1)

    inverted_root3 = solution.invert_tree(root3)
    assert inverted_root3.val == 1
    assert inverted_root3.left is None
    assert inverted_root3.right is None

    print("All test cases passed!")

test_invert_tree()
```

```python
# 4
# Merge Two Binary Trees - Leetcode 617
class NodeTree:

    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def merge_trees(self, tree_one: NodeTree, tree_two: NodeTree) -> NodeTree:

        if tree_one is None and tree_two is None:
            return None
        val_one = tree_one.val if tree_one else 0
        val_two = tree_two.val if tree_two else 0
        root = NodeTree(val_one + val_two)
        root.left = self.merge_trees(tree_one.left if tree_one else None, tree_two.left if tree_two else None)
        root.right = self.merge_trees(tree_one.right if tree_one else None, tree_two.right if tree_two else None)

        return root

def test_merge_trees():
    solution = Solution()

    # Test Case 1: Merging two balanced trees
    tree1 = NodeTree(1)
    tree1.left = NodeTree(3)
    tree1.right = NodeTree(2)
    tree1.left.left = NodeTree(5)

    tree2 = NodeTree(2)
    tree2.left = NodeTree(1)
    tree2.right = NodeTree(3)
    tree2.left.right = NodeTree(4)
    tree2.right.right = NodeTree(7)

    merged_tree = solution.merge_trees(tree1, tree2)
    assert merged_tree.val == 3
    assert merged_tree.left.val == 4
    assert merged_tree.right.val == 5
    assert merged_tree.left.left.val == 5
    assert merged_tree.left.right.val == 4
    assert merged_tree.right.right.val == 7

    # Test Case 2: Merging with one empty tree
    tree3 = NodeTree(1)

    tree4 = None

    merged_tree2 = solution.merge_trees(tree3, tree4)
    assert merged_tree2.val == 1
    assert merged_tree2.left is None
    assert merged_tree2.right is None

    # Test Case 3: Merging with both trees being empty
    tree5 = None

    tree6 = None

    merged_tree3 = solution.merge_trees(tree5, tree6)
    assert merged_tree3 is None

    print("All test cases passed!")

test_merge_trees()
```

```python
# 5
# Convert Sorted Array to Binary Search Tree - Leetcode 108 - Python
class NodeTree:

    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def sorted_array_to_binary_search_tree(self, nums: List[int]) -> NodeTree:

        def helper(left, right):
            if left > right:
                return None
            mid = (left + right) // 2
            root = TreeNode(nums[mid])
            root.left = helper(left, mid - 1)
            root.right = helper(mid + 1, right)
            return root

        return helper(0, len(nums) - 1)

def test_sorted_array_to_binary_search_tree():
    solution = Solution()

    # Test Case 1: Sorted array with odd number of elements
    nums1 = [-10, -3, 0, 5, 9]
    root1 = solution.sorted_array_to_binary_search_tree(nums1)
    assert root1.val == 0
    assert root1.left.val == -10
    assert root1.left.right.val == -3
    assert root1.right.val == 5
    assert root1.right.right.val == 9

    # Test Case 2: Sorted array with even number of elements
    nums2 = [1, 3, 5, 7]
    root2 = solution.sorted_array_to_binary_search_tree(nums2)
    assert root2.val == 3
    assert root2.left.val == 1
    assert root2.right.val == 5
    assert root2.right.right.val == 7

    # Test Case 3: Empty array
    nums3 = []
    root3 = solution.sorted_array_to_binary_search_tree(nums3)
    assert root3 is None

    print("All test cases passed!")

test_sorted_array_to_binary_search_tree()
```

```python
# 6
# Validate Binary Search Tree - Depth First Search - Leetcode 98
class NodeTree:

    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def is_valid(self, root: NodeTree) -> bool:

        def valid(node, left, right):
            if not node:
                return True
            if node.val <= left or node.val >= right:
                return False
            return valid(node.left, left, node.val) and valid(node.right, node.val, right)

        return valid(root, float("-inf"), float("inf"))

def test_is_valid():
    solution = Solution()

    # Test Case 1: A valid binary search tree
    root1 = NodeTree(2)
    root1.left = NodeTree(1)
    root1.right = NodeTree(3)
    assert solution.is_valid(root1) == True

    # Test Case 2: An invalid binary search tree
    root2 = NodeTree(5)
    root2.left = NodeTree(1)
    root2.right = NodeTree(4)
    root2.right.left = NodeTree(3)
    root2.right.right = NodeTree(6)
    assert solution.is_valid(root2) == False

    # Test Case 3: An empty tree
    assert solution.is_valid(None) == True

    # Test Case 4: A tree with a single node
    root3 = NodeTree(1)
    assert solution.is_valid(root3) == True

    # Test Case 5: A tree with two levels
    root4 = NodeTree(2)
    root4.left = NodeTree(1)
    root4.right = NodeTree(3)
    assert solution.is_valid(root4) == True

    # Test Case 6: A tree with three levels
    root5 = NodeTree(10)
    root5.left = NodeTree(5)
    root5.right = NodeTree(15)
    root5.left.left = NodeTree(3)
    root5.left.right = NodeTree(8)
    root5.right.left = NodeTree(12)
    root5.right.right = NodeTree(20)
    assert solution.is_valid(root5) == True

    print("All test cases passed!")

test_is_valid()
```

```python
# 7
# Triangle - Dynamic Programming made Easy - Leetcode 120
class Solution:
    def min_total(self, triangle: List[List[int]]) -> int:
        dp = [0] * (len(triangle) + 1)
        for row in triangle[::-1]:
            for idx, val in enumerate(row):
                dp[idx] = val + min(dp[idx], dp[idx + 1])
        return dp[0]

def test_min_total():
    solution = Solution()

    # Test Case 1: A small triangle
    triangle1 = [
        [2],
        [3, 4],
        [6, 5, 7],
        [4, 1, 8, 3]
    ]
    assert solution.min_total(triangle1) == 11

    # Test Case 2: Another triangle
    triangle2 = [
        [2],
        [3, 4],
        [6, 5, 7],
        [4, 1, 8, 3],
        [7, 8, 9, 6, 10]
    ]
    assert solution.min_total(triangle2) == 19

    # Test Case 3: Triangle with negative values
    triangle3 = [
        [-1],
        [2, 3],
        [1, -1, -3]
    ]
    assert solution.min_total(triangle3) == -1

    # Test Case 4: Empty triangle
    triangle4 = []
    assert solution.min_total(triangle4) == 0

    # Test Case 5: Triangle with a single element
    triangle5 = [[5]]
    assert solution.min_total(triangle5) == 5

    print("All test cases passed!")

test_min_total()
```

```python
# 8
# Sum Root to Leaf Numbers - Coding Interview Question - Leetcode 129
class NodeTree:

    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def sum_numbers(self, root: NodeTree) -> bool:

        def dfs(curr, num):
            if not curr:
                return 0
            num = num * 10 + curr.val
            if not curr.left and not curr.right:
                return num
            return dfs(curr.left, num) + dfs(curr.right, num)

        return dfs(root, 0)

def test_sum_numbers():
    solution = Solution()

    # Test Case 1: A simple binary tree
    root1 = NodeTree(1)
    root1.left = NodeTree(2)
    root1.right = NodeTree(3)
    assert solution.sum_numbers(root1) == 25  # 12 + 13 = 25

    # Test Case 2: A more complex binary tree
    root2 = NodeTree(4)
    root2.left = NodeTree(9)
    root2.right = NodeTree(0)
    root2.left.left = NodeTree(5)
    root2.left.right = NodeTree(1)
    assert solution.sum_numbers(root2) == 1026  # 495 + 491 + 40 = 1026

    # Test Case 3: An empty tree
    assert solution.sum_numbers(None) == 0

    # Test Case 4: A tree with a single node
    root3 = NodeTree(7)
    assert solution.sum_numbers(root3) == 7

    # Test Case 5: A tree with two levels
    root4 = NodeTree(3)
    root4.left = NodeTree(5)
    root4.right = NodeTree(8)
    assert solution.sum_numbers(root4) == 73  # 35 + 38 = 38

    print("All test cases passed!")

test_sum_numbers()
```

```python
# 9
# Unique Binary Search Trees - Leetcode 96 - Python Dynamic Programming
class Solution:

    def number_of_trees(self, n: int) -> int:
        dp = [0] * (n + 1)
        dp[0] = dp[1] = 1

        for nodes in range(2, n + 1):
            total = 0
            for root in range(1, nodes + 1):
                left = root - 1
                right = nodes - root
                total += dp[left] * dp[right]
            dp[nodes] = total

        return dp[n]

def test_number_of_trees():
    solution = Solution()

    # Test Case 1: n = 3
    assert solution.number_of_trees(3) == 5

    # Test Case 2: n = 4
    assert solution.number_of_trees(4) == 14

    # Test Case 3: n = 1
    assert solution.number_of_trees(1) == 1

    # Test Case 4: n = 2
    assert solution.number_of_trees(2) == 2

    print("All test cases passed!")

test_number_of_trees()
```

```python
# 10
# Binary Tree Right Side View - Breadth First Search - Leetcode 199
import collections

class NodeTree:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def sideview(self, root: NodeTree) -> List[int]:
        if not root:
            return []

        res = []
        q = collections.deque([root])

        while q:
            right_side = None
            for i in range(len(q)):
                node = q.popleft()
                if node:
                    right_side = node
                    q.append(node.left)
                    q.append(node.right)
            if right_side:
                res.append(right_side.val)

        return res

def test_sideview():
    solution = Solution()

    # Test Case 1: A binary tree
    root1 = NodeTree(1)
    root1.left = NodeTree(2)
    root1.right = NodeTree(3)
    root1.left.right = NodeTree(5)
    root1.right.right = NodeTree(4)
    assert solution.sideview(root1) == [1, 3, 4]

    # Test Case 2: Another binary tree
    root2 = NodeTree(5)
    root2.left = NodeTree(3)
    root2.right = NodeTree(7)
    root2.left.left = NodeTree(1)
    root2.left.right = NodeTree(4)
    root2.right.right = NodeTree(9)
    assert solution.sideview(root2) == [5, 7, 9]

    # Test Case 3: An empty tree
    assert solution.sideview(None) == []

    print("All test cases passed!")

test_sideview()
```
