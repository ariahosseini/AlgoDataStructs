```python
# Example 1: Finding the factorial of a number
# Example 2: Fibonacci Series
# Example 3: Finding the power of a number
# Example 4: Combination
# Example 5: Tower of Hanoi

class Solution:

    def factorial(self, n: int) -> int:
        if n<=0 or type(n)!=int:
            print("Invalid integer!")
            return None
        elif n == 1:
            return 1
        else:
            return n*self.factorial(n-1)

    def fibonacci(self, idx: int) -> int:
        if idx<0 or type(idx)!=int:
            print("Invalid integer!")
            return None
        if idx == 0 or idx==1:
            return 1
        else:
            return self.fibonacci(idx-2)+self.fibonacci(idx-1)

    def power(self, base: int, exponent: int) -> int:
        if type(base) != int:
            print("Invalid base!")
            return None
        elif type(exponent) != int:
            print("Invalid exponent!")
            return None
        if exponent == 1:
            return base
        else:
            return base*self.power(base,exponent-1)

    def combination(self: int, n, r: int) -> int:
        if type(n) != int or type(r) != int or r>n:
            print("Invalid integers!")
            return None
        return self.factorial(n)/self.factorial(r)/self.factorial(n-r)

    def tower_hanoi(self, n , source, destination, auxiliary):
        if n==1:
            print("Move disk {} from {} to {}".format(n, source, destination))
            return
        self.tower_hanoi(n-1, source, auxiliary, destination)
        print ("Move disk {} from {} to {}".format(n, source, destination))
        self.tower_hanoi(n-1, auxiliary, destination, source)


# test
sol = Solution()
assert sol.factorial(4) == 24
assert sol.fibonacci(4) == 5
assert sol.power(2,6) == 64
assert sol.combination(4,3) == 4
sol.tower_hanoi(n=3, source="A", auxiliary="B", destination="C")
```

```python
# Example 1: Sum of digits of a number
# Example 2: Sum of the first n natural numbers
# Example 3: Least common multiple (LCM) of two numbers
# Example 4: Greatest common divisor (GCD) of two numbers
# Example 5: Reverse a string
# Example 6: Pascals triangle

class Solution:

    def sum_digits(self, num: int) -> int:
        if num == 0:
            return 0
        else:
            return num%10 + sum_digits(num//10)

    def sum_first_n_natural_numbers(self, n: int) -> int:
        if n == 1:
            return 1
        else:
            return n + self.sum_first_n_natural_numbers(n-1)

    def gcd(self, a: int, b: int) -> int:
        if a > b:
            a, b = b, a
        if a == 0:
            return b
        if a == 1:
            return 1
        else:
            return self.gcd(a, b % a)

    def lcm(self, a: int, b: int) -> int:
        remainder = a % b
        if not remainder:
            return a
        else:
            return a*self.lcm(b,remainder)//remainder

    def reverse(self, s: str) -> str:
        if len(s) == 0:
            return s
        else:
            return self.reverse(s[1:]) + s[0]

    def pascal(self, n: int) -> List[int]:
        if n == 1:
            return [1]
        else:
            curr = [1]
            prev = self.pascal(n-1)
            for i in range(len(prev)-1):
                curr.append(prev[i] + prev[i+1])
            curr += [1]
        return curr
# test
sol = Solution()
assert sol.sum_digits(132) == 6
assert sol.sum_first_n_natural_numbers(6) == 21
assert sol.gcd(6, 8) == 2
assert sol.lcm(6, 8) == 24
assert sol.reverse("OpenAI") == "IAnepO"
assert sol.pascal(3) == [1, 2, 1]
```

```python
# Example 1: Sum of odd numbers
# Example 2: Sum of odd and subtraction of even numbers
# Example 3: Palindromes
# Example 4: Remove vowels
# Example 5: Double letters
# Example 6: Check if maze is solvable
# Example 7: Solve a maze
# Example 8: Converting sets to lists in a nested data structure
# Example 9: Least number of coins

class Solution:

    def sum_odd(self, lst: List[int]) -> int:
        if len(lst) == 0:
            return 0
        elif lst[0] % 2 != 0:
            return lst[0] + self.sum_odd(lst[1:])
        else:
            return self.sum_odd(lst[1:])

    def sum_odd_subtrac_even(self, lst: List[int]) -> int:
        if len(lst) == 0:
            return 0
        elif lst[0] % 2 != 0:
            return lst[0] + self.sum_odd_subtrac_even(lst[1:])
        else:
            return -1*lst[0] + self.sum_odd_subtrac_even(lst[1:])

    def is_palindrome(self, string: str) -> bool:
        if len(string) <= 1:
            return True
        elif string[0] == " ":
            return self.is_palindrome(string[1:])
        elif string[-1] == " ":
            return self.is_palindrome(string[:-1])
        elif string[0] == string[-1]:
            return self.is_palindrome(string[1:-1])
        else:
            return False

    def remove_vowels(self, string: str) -> str:
        vowels = "aeiouAEIOU"
        if len(string) == 0:
            return ""
        elif string[0] in vowels:
            return self.remove_vowels(string[1:])
        else:
            return string[0] + self.remove_vowels(string[1:])

    def double(self, string: str) -> str:
        if len(string) == 0:
            return ""
        else:
            return string[0] * 2 + self.double(string[1:])

    def is_maze_solvable(self, maze):
        rows = len(maze)
        cols = len(maze[0])
        visited = [[False] * cols for _ in range(rows)]
        def explore(x, y):
            if x < 0 or x >= rows or y < 0 or y >= cols:
                return False
            if maze[x][y] == 'X':
                return True
            if maze[x][y] == '#' or visited[x][y]:
                return False
            visited[x][y] = True
            result = explore(x + 1, y) or explore(x - 1, y) or explore(x, y + 1) or explore(x, y - 1)
            visited[x][y] = False
            return result
        for i in range(rows):
            for j in range(cols):
                if maze[i][j] == 'S':
                    return explore(i, j)
        return False

    def solve_maze(self, maze):
        rows = len(maze)
        cols = len(maze[0])
        visited = [[False] * cols for _ in range(rows)]
        path = []
        def explore(x, y):
            if x < 0 or x >= rows or y < 0 or y >= cols:
                return False
            if maze[x][y] == 'X':
                path.append('X')
                return True
            if maze[x][y] == '#' or visited[x][y]:
                return False
            visited[x][y] = True
            if explore(x + 1, y):
                path.append('down')
                return True
            elif explore(x - 1, y):
                path.append('up')
                return True
            elif explore(x, y + 1):
                path.append('right')
                return True
            elif explore(x, y - 1):
                path.append('left')
                return True
            visited[x][y] = False
            return False
        for i in range(rows):
            for j in range(cols):
                if maze[i][j] == 'S':
                    if explore(i, j):
                        path.reverse()
                        return ['S']+path
        return []


# test
sol = Solution()
assert sol.sum_odd([1,2,3,4]) == 4
assert sol.sum_odd_subtrac_even([1,2,3,4]) == -2
assert sol.is_palindrome("radar") == True
assert sol.is_palindrome("hello olleh") == True
assert sol.remove_vowels("Hello, World!") == "Hll, Wrld!"
assert sol.double("OpenAI") == "OOppeennAAII"
maze = ['S----',
        '#-#--',
        '#---#',
        '-##-X']
assert sol.is_maze_solvable(maze) == True
assert sol.solve_maze(maze) == ['S', 'right', 'down', 'down', 'right', 'right', 'down', 'right', 'X']
```

```python
# Example 1: Generate subsets
# Example 2: Generate combinations with no replacement
# Example 3: Generate combinations with replacement
# Example 4: Generate permutations
# Example 5: Generate permutations with no replacement
# Example 6: Generate permutations with replacement

class Solution:

    def subsets(self, nums: List[int]) -> List[List[int]]:
        output = [[]]
        n = len(nums)
        for num in nums:
            temp = [subset + [num] for subset in output]
            output.extend(temp)
        return output

    def combinations(self, nums: List[int], k: int) -> List[List[int]]:
        def generate(nums: List[int], k: int, start: int, curr: List[int], final: List[int]):
            if k == 0:
                final.append(curr)
                return None
            for i in range(start, len(nums)):
                generate(nums=nums, k=k-1, start=i+1, curr=curr+[nums[i]], final=final)
        output = []
        empty = []
        generate(nums=nums, k=k, start=0, curr=empty, final=output)
        return output

    def combinations_with_replacement(self, nums: List[int], k: int) -> List[List[int]]:
        def generate(nums: List[int], k: int, start: int, curr: List[int], final: List[int]):
            if k == 0:
                final.append(curr)
                return None
            for i in range(start, len(nums)):
                generate(nums=nums, k=k-1, start=i, curr=curr+[nums[i]], final=final)
        output = []
        empty = []
        generate(nums=nums, k=k, start=0, curr=empty, final=output)
        return output

    def permutations(self, nums: List[int]) -> List[List[int]]:
        def generate(nums: List[int], start: int, final: List[int]):
            if start == len(nums)-1:
                final.append(nums.copy())
                return None
            for i in range(start, len(nums)):
                nums[start], nums[i] = nums[i], nums[start]
                generate(nums=nums, start=start+1, final=final)
                nums[start], nums[i] = nums[i], nums[start]
        output = []
        start = 0
        generate(nums=nums, start=start, final=output)
        return output

    def permutations_of_len_k(self, nums: List[int], k: int) -> List[List[int]]:
        def generate(nums: List[int], k: int, curr: List[int], final: List[int]):
            if k == 0:
                final.append(curr)
                return
            for i in range(len(nums)):
                digit = nums[i]
                remain_digits = nums[:i] + nums[i+1:]
                generate(nums=remain_digits, k=k-1, curr=curr+[digit], final=final)
        output = []
        empty = []
        generate(nums=nums, k=k, curr=empty, final=output)
        return output

    def permutations_of_len_k_with_replacement(self, nums: List[int], k: int) -> List[List[int]]:
        def generate(nums: List[int], k: int, curr: List[int], final: List[int]):
            if k == 0:
                final.append(curr)
                return
            for i in range(len(nums)):
                digit = nums[i]
                generate(nums=nums, k=k-1, curr=curr+[digit], final=final)
        output = []
        empty = []
        generate(nums=nums, k=k, curr=empty, final=output)
        return output

# test
sol = Solution()
sol.subsets([1,2,3,4])
sol.combinations([1,2,3], 2)
sol.combinations_with_replacement([1,2,3], 2)
sol.permutations([1,2,3])
sol.permutations_of_len_k([1,2,3],2)
sol.permutations_of_len_k_with_replacement([1,2,3],2)
```
