```python
class Node:

    def __init__(self, data):
        self.data = data
        self.nref = None
        self.pref = None

class DoublyLinkedList:

    def __init__(self):
        self.head = None

    def forward_display(self):
        n = self.head
        if not n:
            print("The linked list is empty!")
            return None
        while n:
            print(n.data, "->", end=" ") if n.nref else print(n.data)
            n = n.nref
        return None

    def backward_display(self):
        n = self.head
        if not n:
            print("The linked list is empty!")
            return None
        while n.nref:
            n = n.nref
        p = n
        while p:
            print(p.data, "->", end=" ") if p.pref else print(p.data)
            p = p.pref
        return None

    def insert_empty(self, data):
        n = self.head
        if n:
            print("The linked list is not empty!")
            return None
        new_node = Node(data)
        self.head = new_node
        return None

    def get_size(self):
        n = self.head
        count = 0
        while n:
            count += 1
            n = n.nref
        return count

    def max_value(self):
        n = self.head
        if not n:
            print("The linked list is empty!")
            return None
        if not n.nref:
            return n.data
        else:
            max_value = n.data
            while n:
                max_value = max(n.data, max_value)
                n = n.nref
        return max_value

    def min_value(self):
        n = self.head
        if not n:
            print("The linked list is empty!")
            return None
        if not n.nref:
            return n.data
        else:
            min_value = n.data
            while n:
                min_value = min(n.data, min_value)
                n = n.nref
        return min_value

    def get_index(self, x):
        n = self.head
        if n is None:
            print("Linked list is empty!")
        index = 0
        while n:
            if n.data==x:
                return index
            index += 1
            n = n.nref
        print("{} does not exist in the linked list!".format(x))
        return None

    def add_begin(self, data):
        n = self.head
        if not n:
            return self.insert_empty(data)
        new_node = Node(data)
        new_node.nref = self.head
        self.head.pref = new_node
        self.head = new_node
        return None

    def add_end(self, data):
        n = self.head
        if not n:
            return self.insert_empty(data)
        if not n.nref:
            return self.add_begin(data)
        while n.nref:
            n = n.nref
        new_node = Node(data)
        new_node.pref = n
        n.nref = new_node
        return None

    def add_after(self, data, x):
        n = self.head
        if n is None:
            print("Linked list is empty!")
            return None
        while n.nref and n.data != x:
            n = n.nref
        if n.data == x and not n.nref:
            return self.add_end(data)
        elif n.data != x and not n.nref:
            print("{} does not exist!".format(x))
            return None
        else:
            new_node = Node(data)
            new_node.pref = n
            new_node.nref = n.nref
            n.nref.pref = new_node
            n.nref = new_node
        return None

    def add_before(self, data, x):
        n = self.head
        if n is None:
            print("Linked list is empty!")
            return None
        if n.data == x:
            return self.add_begin(data)
        while n.nref and n.nref.data != x:
            n = n.nref
        if not n.nref:
            print("{} does not exist!".format(x))
            return None
        return self.add_after(data, n.data)

    def delete_begin(self):
        if self.head is None:
            print("Linked list is already empty!")
            return None
        else:
            n = self.head
            n.nref.pref = None
            self.head = self.head.nref
        return None

    def delete_end(self):
        if self.head is None:
            print("Linked list is already empty!")
            return None
        else:
            n = self.head
            if n.nref is None:
                return self.delete_begin()
            else:
                while n.nref.nref is not None:
                    n = n.nref
                n.nref = None
        return None

    def delete_by_value(self, x):
        if self.head is None:
            print("Linked list is empty!")
            return None
        n = self.head
        if n.data == x:
            return self.delete_begin()
        while n.nref is not None:
            if n.data == x:
                n.nref.pref = n.pref
                n.pref.nref = n.nref
                return None
            n = n.nref
        if n.nref is None:
            return self.delete_end() if n.data == x else print("{} does not exist in the linked list!".format(x))




LL = DoublyLinkedList()
LL.insert_empty(10)
LL.add_begin(5)
LL.add_begin(0)
LL.add_end(15)
LL.add_after(20, 5)
LL.add_after(30, 15)
LL.add_before(-10, 30)
LL.delete_begin()
LL.delete_end()
LL.delete_by_value(20)
LL.forward_display()
LL.backward_display()
print("Length of the linked list is {}".format(LL.get_size()))
print("Maximum value in the linked list is {}".format(LL.max_value()))
print("Minimum value in the linked list is {}".format(LL.min_value()))
```
